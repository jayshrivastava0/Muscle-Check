```python
import numpy as np
import matplotlib.pyplot as plt

# Create original grid
x, y = np.meshgrid(np.linspace(0, 10, 11), np.linspace(0, 10, 11))

# Distorted grid (simulating muscle pump)
x_distorted = x + 0.3 * np.sin(y / 2)
y_distorted = y + 0.3 * np.sin(x / 2)

# Plotting
plt.figure(figsize=(12, 6))

# Before Workout (Normal Muscle)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-', alpha=0.5)
plt.plot(x.T, y.T, 'b-', alpha=0.5)
plt.scatter(x, y, color='blue', marker='x', label='Original Grid')
plt.title("Before Workout (Normal Muscle)")
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.legend()

# After Workout (Muscle Pump)
plt.subplot(1, 2, 2)
plt.plot(x, y, 'gray', alpha=0.3, linestyle='--')  # Ghost grid
plt.plot(x.T, y.T, 'gray', alpha=0.3, linestyle='--')  # Ghost grid
plt.plot(x_distorted, y_distorted, 'r-', alpha=0.5)
plt.plot(x_distorted.T, y_distorted.T, 'r-', alpha=0.5)
plt.scatter(x_distorted, y_distorted, color='red', marker='x', label='Expanded Grid')
plt.title("After Workout (Muscle Pump)")
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.legend()

plt.tight_layout()
plt.show()
```
